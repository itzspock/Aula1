```python
pasta_destino = 'drive/MyDrive/WebScraping/'

if not os.path.exists(pasta_destino):
    os.makedirs(pasta_destino)

for pdf_url in pdf_links:
    try:
        response = requests.get(pdf_url)
        if response.status_code == 200:
            nome_arquivo = pdf_url.split('/')[-1]

            destino = os.path.join(pasta_destino, nome_arquivo)

            with open(destino, 'wb') as pdf_file:
                pdf_file.write(response.content)

            print(f"PDF baixado com sucesso. Salvo em: {destino}")
        else:
            print(f"Falha ao baixar o PDF. Código de status: {response.status_code}")

    except Exception as e:
        print(f"Erro ao baixar o PDF: {e}")
```

```python
def extract_used_cars_data(text):
    lines = text.splitlines()
    pattern = r'(\d+º)\s+([\w\s-]+)\s+([\d.]+)'
    data = re.findall(pattern, "\n".join(lines))
    return data[:50]

lista_arquivos = os.listdir(pasta_destino)

all_data = []

for nome_arquivo in lista_arquivos:
    if nome_arquivo.endswith('.pdf'):
        caminho_pdf = os.path.join(pasta_destino, nome_arquivo)

        pdf_document = fitz.open(caminho_pdf)

        todo_texto = ""
        for num_pagina in range(pdf_document.page_count):
            pagina = pdf_document[num_pagina]
            todo_texto = todo_texto + pagina.get_text("text")

        pdf_document.close()

        data = extract_used_cars_data(todo_texto)

        if data is not None:
            mes_ano = re.search(r'(\d{4}_\d{2})', nome_arquivo).group(1)
            for i, dados_entrada in enumerate(data, start=1):
                lista_entrada = list(dados_entrada)
                lista_entrada[0] = f"{i}º"
                lista_entrada.append(mes_ano)
                all_data.append(lista_entrada)

all_data
```
